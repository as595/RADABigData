```python
filename = "MSX6C_filtered.txt"
infile = open(filename,'r')
```

```python
import numpy as np

F8=[];F12=[];F14=[];F21=[]
while True:
    line = infile.readline()
    items=line.split()
    
    if not line: break
    
    if len(items)>0:
        
        if items[0][0]=='G':

            if items[5]=='0':
                if items[7]=='0':
                    F8.append(float(items[8]))
                    F12.append(float(items[11]))
                    
                    if items[12]=='0':
                        F14.append(float(items[13]))
                        F21.append(float(items[16]))                        
                    else:
                        F14.append(float(items[14]))
                        F21.append(float(items[17]))
                    
                else:
                    F8.append(float(items[9]))
                    F12.append(float(items[12]))

                    if items[13]=='0':
                        F14.append(float(items[14]))
                        F21.append(float(items[17]))                        
                    else:
                        F14.append(float(items[15]))
                        F21.append(float(items[18]))

            else:
                if items[8]=='0':
                    F8.append(float(items[9]))
                    F12.append(float(items[12]))

                    if items[13]=='0':
                        F14.append(float(items[14]))
                        F21.append(float(items[17]))                        
                    else:
                        F14.append(float(items[15]))
                        F21.append(float(items[18]))

                else:
                    F8.append(float(items[10]))
                    F12.append(float(items[13]))
                    
                    if items[13]=='0':
                        F14.append(float(items[15]))
                        F21.append(float(items[18]))                        
                    else:
                        F14.append(float(items[16]))
                        F21.append(float(items[19]))


F8=np.array(F8)
F12=np.array(F12)
F14=np.array(F14)
F21=np.array(F21)  
```

```python
import pylab as pl

pl.subplot(111)
pl.scatter(F21/F14,F14/F8,s=1)
pl.xlabel("F21/F14")
pl.ylabel("F14/F8")
pl.axis([0.1,30,0.1,30])
pl.loglog()
pl.show()
```
